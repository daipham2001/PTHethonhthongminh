```python
# Load dữ liệu
import numpy as np
import pandas as pd
df = pd.read_csv('diabetes.csv')
df.info()
```

```python
# Kiểm tra các giá trị null
print("Nulls")
print("=====")
print(df.isnull().sum())
```

```python
# Kiểm tra các giá trị 0
print("0s")
print("==")
print(df.eq(0).sum())
```

```python
# Thay thế các giá trị 0 bằng NaN
df[['Glucose','BloodPressure','SkinThickness',
 'Insulin','BMI','DiabetesPedigreeFunction','Age']] = \
df[['Glucose','BloodPressure','SkinThickness',
 'Insulin','BMI','DiabetesPedigreeFunction','Age']].replace(0,np.NaN)
```

```python
# Thay các giá trị NaN bằng giá trị trung bình của mỗi cột
df.fillna(df.mean(), inplace = True)
```

```python
# Kiểm tra lại các giá trị 0 xem có còn hay không? 
print(df.eq(0).sum())
```

```python
# Kiểm tra mối tương quan giữa các tính năng
# Kết quả sau đây cho thấy mức đường huyết của một bệnh nhân sau thử nghiệm dung nạp đường uống trong 
# 2 giờ có ít mối quan hệ với số lần mang thai của một bệnh nhân (0,127911), 
# nhưng nó có mối quan hệ đáng kể với kết quả (0,492928)
corr = df.corr()
print(corr)
```

```python
# Biểu đồ hiển thị sự tương quan 
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1) 
# vẽ biểu đồ kết quả được trả về bởi hàm corr() dưới dạng ma trận

fig.colorbar(cax)
ticks = np.arange(0,len(df.columns),1)
ax.set_xticks(ticks)

ax.set_xticklabels(df.columns)
plt.xticks(rotation = 90)

# ax.set_yticklabels(df.columns)
ax.set_yticks(ticks)

#---print the correlation factor---
for i in range(df.shape[1]):
    for j in range(9):
        text = ax.text(j, i, round(corr.iloc[i][j],2),ha="center", va="center", color="w")
plt.show()
#Các hình khối có màu sắc gần nhất với màu đỏ phản ánh các yếu tố tương quan cao nhất, 
#trong khi các hình khối gần nhất với màu xanh lam đại diện cho các yếu tố tương quan thấp nhất.
```

```python
# Một cách khác để vẽ ma trận tương quan là sử dụng hàm Seaborn’s heatmap ()
import seaborn as sns
sns.heatmap(df.corr(),annot=True)
# lấy tham chiếu đến con số hiện tại và đặt kích thước của nó
fig = plt.gcf()
fig.set_size_inches(8,8)
```

```python
# lấy bốn tính năng hàng đầu có mối tương quan cao nhất
print(df.corr().nlargest(4, 'Outcome').index)
```

```python
# in ra 4 giá trị có mối tương quan cao nhất
print(df.corr().nlargest(4, 'Outcome').values[:,8])
```

```python
# Đánh giá thuật toán(Evaluating the Algorithms)
# Hồi quy tuyến tính(Logistic Regression)
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
#---features---
X = df[['Glucose','BMI','Age']]
#---label---
y = df.iloc[:,8]
log_regress = linear_model.LogisticRegression()
# xác thực chéo 10 lần để có được điểm trung bình của thuật toán
log_regress_score = cross_val_score(log_regress, X, y, cv=10, scoring='accuracy').mean()
print(log_regress_score)
```

```python
#lưu kết quả đào tạo mô hình vào mảng để so sánh với các thuật toán khác
result = []
result.append(log_regress_score)
```

```python
# K-Nearest Neighbors(KNN)
from sklearn.neighbors import KNeighborsClassifier
# tạo danh sách cv trống để cho chứa điểm cv
cv_scores = []
folds = 10 # số lần gấp
# tạo danh sách ks có k là số lẻ
ks = list(range(1,int(len(X) * ((folds - 1)/folds)), 2))
# thực hiện xác thực chéo k lần
for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, X, y, cv=folds, scoring='accuracy').mean()
    cv_scores.append(score)
# lấy giá trị điểm max của danh sách cv
knn_score = max(cv_scores)
# tìm k tối ưu 
optimal_k = ks[cv_scores.index(knn_score)]
print(f"The optimal number of neighbors is {optimal_k}")
print(knn_score)
# tìm k xong rồi thì lưu kết quả vào mảng để so sánh với các thuật toán khác
result.append(knn_score)
```

```python
# Support Vector Machines
# Sử dụng 2 nhãn cho thuật toán Support Vector Machines là: linear và rbf
from sklearn import svm
linear_svm = svm.SVC(kernel='linear')
linear_svm_score = cross_val_score(linear_svm, X, y,cv=10, scoring='accuracy').mean()
print(linear_svm_score)
result.append(linear_svm_score)
```

```python
# Dùng nhãn rbf
rbf = svm.SVC(kernel='rbf')
rbf_score = cross_val_score(rbf, X, y, cv=10, scoring='accuracy').mean()
print(rbf_score)
result.append(rbf_score)
```

```python
# Chọn thuật toán hoạt động tốt nhất mà chúng ta đã sử dụng 4 thuật toán khác nhau
algorithms = ["Logistic Regression", "K Nearest Neighbors", "SVM Linear Kernel", "SVM RBF Kernel"]
cv_mean = pd.DataFrame(result,index = algorithms)
cv_mean.columns=["Accuracy"]
cv_mean.sort_values(by="Accuracy",ascending=False)
```

```python
# Training and Saving the Model
# thuật toán hoạt động tốt nhất cho tập dữ liệu của chúng tôi là KNN với k = 19
knn = KNeighborsClassifier(n_neighbors=19)
knn.fit(X, y)
```

```python
import pickle
# lưu mô hình 
filename = 'diabetes.sav'
# ghi vào tệp 
pickle.dump(knn, open(filename, 'wb'))
```

```python
# Load mô hình từ file
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
Glucose = 65
BMI = 70
Age = 50
prediction = loaded_model.predict([[Glucose, BMI, Age]])
print(prediction)
if (prediction[0]==0):
    print("Non-diabetic")
else:
    print("Diabetic")
```

```python
# Chuyển về tỷ lệ %
proba = loaded_model.predict_proba([[Glucose, BMI, Age]])
print(proba)
print("Confidence: " + str(round(np.amax(proba[0]) * 100 ,2)) + "%")
```

```python
# Deploying the Model
import pickle
from flask import Flask, request, json, jsonify
import numpy as np
app = Flask(__name__)
#---the filename of the saved model---
filename = 'diabetes.sav'
#---load the saved model---
loaded_model = pickle.load(open(filename, 'rb'))
@app.route('/diabetes/v1/predict', methods=['POST'])
def predict():
    #---get the features to predict---
    features = request.json
    #---create the features list for prediction---
    features_list = [features["Glucose"],
    features["BMI"],
    features["Age"]]
    #---get the prediction class---
    prediction = loaded_model.predict([features_list])
    #---get the prediction probabilities---
    confidence = loaded_model.predict_proba([features_list])
    #---formulate the response to return to client---
    response = {}
    response['prediction'] = int(prediction[0])
    response['confidence'] = str(round(np.amax(confidence[0]) * 100 ,2))
    return jsonify(response)
if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
